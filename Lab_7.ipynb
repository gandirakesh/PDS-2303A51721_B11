```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("daily_temperature.csv")  # columns: Day, Temperature

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(df["Day"], df["Temperature"], marker='o', linestyle='-', linewidth=2)

# Labels and title
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Changes (Weather Dataset)")
plt.grid(True)

plt.show()

# Interpretation
hottest_day = df.loc[df["Temperature"].idxmax()]
coldest_day = df.loc[df["Temperature"].idxmin()]

print("Hottest Day:", hottest_day["Day"], "with", hottest_day["Temperature"], "°C")
print("Coldest Day:", coldest_day["Day"], "with", coldest_day["Temperature"], "°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")  # columns: Quarter, Sales

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Quarter"], df["Sales"], marker='o', linestyle='-', linewidth=2, label="Sales")

# Highlight peak sales
peak = df.loc[df["Sales"].idxmax()]
plt.plot(peak["Quarter"], peak["Sales"], marker='o', color='red', markersize=12, label="Peak Sales")

# Labels and title
plt.xlabel("Quarter")
plt.ylabel("Sales (in $)")
plt.title("Quarterly Sales Performance (Retail Sales Dataset)")
plt.grid(True)
plt.legend()

plt.show()

# Discussion
print("Peak Sales occurred in", peak["Quarter"], "with value:", peak["Sales"])
print("\nDiscussion:")
print("The sales data shows how performance varies across quarters.")
print("If sales peak in Q4, it may indicate holiday/seasonal shopping boosts.")
print("Lower sales in Q1 or Q2 could reflect post-holiday slowdown.")
```
